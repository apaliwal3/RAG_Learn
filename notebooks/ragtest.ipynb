```python
# Text file
import os
os.makedirs("../data/textfiles", exist_ok = True)
```

```python
sample_texts = {
  "../data/textfiles/python_intro.txt":""" Python is a high-level, versatile programming language known for its simplicity and readability. Designed with a clean and intuitive syntax, it allows developers to express complex ideas in fewer lines of code compared to many other languages. This makes it an excellent choice for beginners, while still being powerful enough for experienced engineers working on large-scale systems.
One of Python’s biggest strengths is its broad range of applications. It is widely used in web development, data analysis, artificial intelligence, machine learning, automation, and scientific computing. Libraries such as NumPy, pandas, and TensorFlow extend its capabilities, enabling developers to build everything from simple scripts to advanced AI models.
Python also has a strong and active community, which contributes to a vast ecosystem of open-source tools and resources. This means that solutions to common problems are often readily available, speeding up development and reducing the need to build everything from scratch.
Overall, Python stands out as a practical and efficient language that balances ease of use with powerful functionality, making it one of the most popular programming languages in the world today.""",
  "../data/textfiles/ml_intro.txt":"""Machine learning (ML) is a field of computer science focused on building systems that can learn from data and improve their performance over time without being explicitly programmed for every task. Instead of relying on fixed rules, ML models identify patterns in data and use those patterns to make predictions or decisions.

At a high level, machine learning can be divided into a few main categories. Supervised learning involves training a model on labeled data, where the correct outputs are known—common tasks include classification and regression. Unsupervised learning works with unlabeled data, aiming to discover hidden structures such as clusters or relationships. Reinforcement learning takes a different approach, where an agent learns by interacting with an environment and receiving feedback in the form of rewards or penalties.

ML powers many of the technologies people use every day. Recommendation systems suggest what to watch or buy next, fraud detection systems flag unusual transactions, and natural language processing enables chatbots and voice assistants to understand human language. In fields like healthcare, finance, and transportation, machine learning is increasingly used to improve decision-making and efficiency.
"""
}

for filepath, content in sample_texts.items():
  with open(filepath, 'w', encoding="utf-8") as f:
    f.write(content)
  
print("Sample files created!")
```

```python
### TextLoader
from langchain_community.document_loaders import TextLoader

loader = TextLoader("../data/textfiles/python_intro.txt", encoding="utf-8")
document=loader.load()
print(document)
```

```python
### Directory Loader
from langchain_community.document_loaders import DirectoryLoader

## load all text files from the directory
dir_loader=DirectoryLoader(
  "../data/textfiles",
  glob="**/*.txt",
  loader_cls= TextLoader,
  loader_kwargs={'encoding': 'utf-8'},
  show_progress=False
)

documents=dir_loader.load()
documents

```

```python
from langchain_community.document_loaders import PyPDFLoader, PyMuPDFLoader
dir_loader = DirectoryLoader(
  "../data/pdf",
  glob="**/*.pdf",
  loader_cls=PyMuPDFLoader,
  show_progress=False
)

pdf_documents=dir_loader.load()
pdf_documents
```
